# Hello Graph World

Load networkx and matplotlib libraries  

```python
import networkx as net
import matplotlib.pyplot as plot
```

We will create a graph called 'o' and add edges to it. 


```python
o=net.Graph()

o.add_edge('Andre','Carol')
o.add_edge('Andre','Beverly')
o.add_edge('Andre','Diane')
o.add_edge('Andre','Fernando')

o.add_edge('Beverly','Ed')
o.add_edge('Beverly','Diane')
o.add_edge('Beverly','Garth')


o.add_edge('Carol','Fernando')
o.add_edge('Carol','Diane')
 
o.add_edge('Diane','Fernando')
o.add_edge('Diane','Garth')
o.add_edge('Diane','Ed')
o.add_edge('Ed','Garth')    


o.add_edge('Fernando','Garth')
o.add_edge('Fernando','Heather')

o.add_edge('Garth','Heather')
 
o.add_edge('Heather','Ike')
o.add_edge('Ike','Jane')

```

Draw the graph. 

```python
net.draw(o, with_labels = True,pos=net.spring_layout(o))
 
```

We could also use 'krackhardt_kite_graph' to draw the graph. 
Andre=0, Beverley=1, Carol=2, Diane=3 Ed=4, Fernando=5, Garth=6, Heather=7, Ike=8, Jane=9

```python
G = net.krackhardt_kite_graph()
net.draw_networkx(G)
 
```
